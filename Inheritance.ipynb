```python
class Student: 
   def __init__ (self,marks):
        self.marks=marks
   def __add__ (self,other):
        return self.marks+other.marks

s1= Student (80)
s2= Student (75)
print(s1+s2)
```

```python
class Parent:
   def func1(self):
     print("This function is in child class.")
class child(Parent):
  def func2(self):
     print ("This funtion is in child class.")
object=child()
object.func1()
object.func2()
```
